## Plot_Ratio_ofLandAreas_onTwoHemishperes_0-400Ma

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

data_Matthews2016 = pd.read_csv('/Users/wenchaocao/Research/41_Paleobiology_Paleotectonics/TotalNumber_Points_onTwoHemispheres_Matthews2016.csv')
data_Scotese2008 = pd.read_csv('/Users/wenchaocao/Research/41_Paleobiology_Paleotectonics/TotalNumber_Points_onTwoHemispheres_Scotese2008.csv')

plt.figure(figsize=(10,6),tight_layout=True) #figsize=(12,8)

#print data[0:1]
plt.plot(data_Matthews2016[0:1].T, 100*data_Matthews2016[5:6].T,'r-',data_Matthews2016[0:1].T, 100*data_Matthews2016[6:7].T,'b-')
plt.scatter(data_Matthews2016[0:1].T, 100*data_Matthews2016[5:6].T, c='red',alpha=0.8, edgecolors='none',s=40)
plt.scatter(data_Matthews2016[0:1].T, 100*data_Matthews2016[6:7].T, c='blue',alpha=0.8, edgecolors='none',s=40)

# Scotese
plt.plot(data_Scotese2008[0:1].T, 100*data_Scotese2008[5:6].T,'g--',data_Scotese2008[0:1].T, 100*data_Scotese2008[6:7].T,'m--')
plt.scatter(data_Scotese2008[0:1].T, 100*data_Scotese2008[5:6].T, c='g',alpha=0.8, edgecolors='none',s=40)
plt.scatter(data_Scotese2008[0:1].T, 100*data_Scotese2008[6:7].T, c='m',alpha=0.8, edgecolors='none',s=40)

plt.legend(('North hemisphere (Mattthews et al., 2016)','South hemisphere (Mattthews et al., 2016)',
            'North hemisphere (Scotese, 2008)','South hemisphere (Scotese, 2008)'),
           bbox_to_anchor=(.6,0.97), loc=1, borderaxespad=0.)
plt.xlabel('Time (Ma)', fontsize=20)
plt.xticks(np.arange(0,401,20), size=10) #, rotation=50
plt.ylabel('Ratio of land area [% global surface]', fontsize=20)
plt.xlim((0,400))
plt.ylim((0,90))
plt.grid()
plt.show
#plt.savefig('Ratio_LandAreas_OnTwoHemispheres_0-400Ma.pdf',dpi=600)
```
