```python
import pygplates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

%matplotlib inline


# sum of coals, evaporites and glacial
Data = pd.read_excel('/Users/wenchaocao/Research/41_Paleobiology_Paleotectonics/2_2_Sum_LithData_EachMap_410-0Ma/Sum_LithData_Matthews2106_EachMap_410-0Ma.xlsx')

Data_bins = pd.read_excel('/Users/wenchaocao/Research/41_Paleobiology_Paleotectonics/3_LithData_Number_InEachBand_SamplingBiasRemoved/Sum_Matthews2016_BinsHavingData_BinningSize5.xlsx')
```
